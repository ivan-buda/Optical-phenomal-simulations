```python
"""
Script to simulate the distribution of photons in a 2D plane and plot the resulting distribution.

Libraries used:
    - numpy
    - matplotlib
    - scipy

Functions defined:
    - None

Variables defined:
    - Nphotons: integer. Number of photons to simulate.
    - xdata, ydata: numpy arrays of size Nphotons x 1. Contain the x and y coordinates of the simulated photons.
    - bins: numpy array. Defines the bins for the histogram.
    - I: 2D numpy array. Contains the counts for each bin in the histogram.
    - X, Y: 2D numpy arrays. Contain the coordinates of each bin in the histogram.
"""

# import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# define the number of photons to simulate
Nphotons = 2**24

# simulate the distribution of photons in a 2D plane
xdata, ydata = np.random.normal(size=(2, Nphotons))

# plot the resulting distribution
plt.figure()
plt.plot(xdata, ydata, '.', color='w', markersize=0.05)
plt.title("Photons location")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_facecolor('k')
plt.show()

# create a histogram of the photon distribution
bins = np.arange(-5, 5, 0.01)
I, _, _ = np.histogram2d(xdata, ydata, bins=[bins, bins])
X, Y = np.meshgrid(bins, bins)

# plot the histogram
plt.figure()
plt.pcolormesh(X, Y, I, shading='flat', cmap='hot')
plt.title("Gaussian distribution")
plt.xlabel("x")
plt.ylabel("y")
plt.plot()
plt.show()
```
